```python
import requests
import json

API_URL = "https://www.air-port-codes.com/api/v1/multi"
API_KEY = "d4ad68a761"

IATA_FALLBACK = {
    "Madrid": "MAD",
    "Quito": "UIO",
    "Bogota": "BOG",
    "Barcelona": "BCN",
    "Frankfurt": "FRA",
    "Berlin": "BER",
    "Roma": "FCO",
    "Sevilla": "SVQ"
}
```

```python
def obtener_iata(ciudad):
    headers = {
        "Content-Type": "application/json",
        "APC-Auth": API_KEY
    }

    payload = {"cities": [ciudad]}

    try:
        response = requests.post(API_URL, headers=headers, json=payload, timeout=10)

        if response.status_code != 200:
            return IATA_FALLBACK.get(ciudad, "N/A")

        data = response.json()

        if "cities" in data and ciudad in data["cities"] and data["cities"][ciudad]:
            return data["cities"][ciudad][0].get("iata", "N/A")

    except Exception:
        pass

    return IATA_FALLBACK.get(ciudad, "N/A")
```

```python
ciudades = ["Madrid", "Quito", "Bogota", "Barcelona", "Frankfurt"]

for ciudad in ciudades:
    print(f"{ciudad} → {obtener_iata(ciudad)}")
```
